```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('avocado.csv')
```

```python
df.head(3)
```

```python
df.info()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.head(3)
```

```python
plt.plot(df['date'],df['average_price'])

plt.show()
```

```python
dfgrup=df.groupby(by="date").mean()
```

```python
dfgrup.head(3)
```

```python
plt.figure(figsize=(14,6))
plt.plot(dfgrup.index,dfgrup['average_price'])
plt.title("Avocado Fiyat Grafiği",fontdict={'fontsize':16})
plt.xlabel("Tarih")
plt.ylabel("Fiyat(USD)")
plt.show()
```

```python
coffee=pd.read_csv('coffee.csv')
```

```python
coffee.tail(3)
```

```python
coffee['Date']=pd.to_datetime(coffee['Date'])
```

```python
coffee['price']=(coffee['Close']/25)*0.7
```

```python
coffee.head(3)
```

```python
coffee=coffee.set_index("Date")
```

```python
coffee.head(3)
```

```python
dfgrup.head(10)
```

```python
coffee=coffee.reindex(pd.date_range("2000-01-03","2022-09-02",freq="D"))
```

```python
coffee=coffee[['price']]
```

```python
coffee.head(10)
```

```python
coffee["price_cmt"]=coffee.shift(1)
```

```python
coffee.head(10)
```

```python
coffee["price_pzr"]=coffee['price_cmt'].shift(1)
```

```python
coffee.head(10)
```

```python
coffee['price']=coffee['price'].fillna(0)
```

```python
coffee['price']=np.where(coffee['price']==0,coffee['price_cmt'],coffee['price'])
```

```python
coffee['price']=coffee['price'].fillna(0)
coffee['price']=np.where(coffee['price']==0,coffee['price_pzr'],coffee['price'])
```

```python
coffee.head(10)
```

```python
cof=coffee.loc[dfgrup.index]
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(14,6))
plt.plot(dfgrup.index,dfgrup['average_price'])
plt.plot(dfgrup.index,cof['price'])
plt.xlabel("Tarih")
plt.ylabel("Fiyat USD")
plt.title("Kahve ve Avocado Fiyatı",fontdict={'fontsize':16})
plt.legend(["Avocado","Kahve"])
plt.savefig("KAF.png")
plt.show()
```
